```python
def scrape_title(webpage_content):
    title = webpage_content.find('h2')
    return title

def scrape_tags(webpage_content):
    tags_box = webpage_content.find('div', class_="tags-box")
    tags = tags_box.find_all('a', class_='tag')
    return tags

def scrape_across_pages(webpage_content, browser):
    quotes = webpage_content.find_all('span', class_='text')
    for x in range(1, 6):
        for quote in quotes:
            print('page:', x, '----------')
            print(quote.text)
        browser.links.find_by_partial_text('Next')
```

```python
url = "http://quotes.toscrape.com"
browser = init_splinter(url)
webpage_content = get_web_content(browser)
scrape_title(webpage_content).text

# scrape all tags
response = scrape_tags(webpage_content)
for tag in response:
    word = tag.text
    print(word)

# scrape across pages
scrape_across_pages(webpage_content, browser)
browser.quit()  
```
